```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import os
```

```python
def plota_fig(df, value, index, func, ylabel, xlabel, opcao='nada'):
    if opcao=='nada':
        pd.pivot_table(df, values=value, index=index, aggfunc=func).plot(figsize=[15,5])
    elif opcao=='unstack':
        pd.pivot_table(df, values=value, index=index, aggfunc=func).unstack().plot(figsize=[15,5])
    elif opcao=='sort':
        pd.pivot_table(df, values=value, index=index, aggfunc=func).sort_values(value).plot(figsize=[15,5])
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    return None

sinasc = pd.read_csv('./input/SINASC_RO_2019_MAR.csv')

max_data = sinasc.DTNASC.max()[:7]
print(max_data)

os.makedirs('./output/figs/'+max_data)

plota_fig(sinasc,'IDADEMAE','DTNASC','mean','mdia idade me por data', 'data nascimento')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por data')

plota_fig(sinasc,'PESO','ESCMAE','median','peso mediano beb', 'escolaridade me', 'sort')
plt.savefig('./output/figs/'+max_data+'/peso mediano por escolaridade me')

plota_fig(sinasc,'IDADEMAE',['DTNASC','SEXO'],'mean','mdia idade me', 'data nascimento', 'unstack')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por sexo')

plota_fig(sinasc,'PESO',['DTNASC','SEXO'],'mean','mdia peso beb', 'data nascimento', 'unstack')
plt.savefig('./output/figs/'+max_data+'/mdia peso beb por sexo')

plota_fig(sinasc,'APGAR1','GESTACAO','mean','apgard mdio', 'gestao', 'sort')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por data')

sinasc = pd.read_csv('./input/SINASC_RO_2019_ABR.csv')

max_data = sinasc.DTNASC.max()[:7]
print(max_data)

os.makedirs('./output/figs/'+max_data)

plota_fig(sinasc,'IDADEMAE','DTNASC','mean','mdia idade me por data', 'data nascimento')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por data')

plota_fig(sinasc,'PESO','ESCMAE','median','peso mediano beb', 'escolaridade me', 'sort')
plt.savefig('./output/figs/'+max_data+'/peso mediano por escolaridade me')

plota_fig(sinasc,'IDADEMAE',['DTNASC','SEXO'],'mean','mdia idade me', 'data nascimento', 'unstack')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por sexo')

plota_fig(sinasc,'PESO',['DTNASC','SEXO'],'mean','mdia peso beb', 'data nascimento', 'unstack')
plt.savefig('./output/figs/'+max_data+'/mdia peso beb por sexo')

plota_fig(sinasc,'APGAR1','GESTACAO','mean','apgard mdio', 'gestao', 'sort')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por data')

sinasc = pd.read_csv('./input/SINASC_RO_2019_MAI.csv')

max_data = sinasc.DTNASC.max()[:7]
print(max_data)

os.makedirs('./output/figs/'+max_data)

plota_fig(sinasc,'IDADEMAE','DTNASC','mean','mdia idade me por data', 'data nascimento')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por data')

plota_fig(sinasc,'PESO','ESCMAE','median','peso mediano beb', 'escolaridade me', 'sort')
plt.savefig('./output/figs/'+max_data+'/peso mediano por escolaridade me')

plota_fig(sinasc,'IDADEMAE',['DTNASC','SEXO'],'mean','mdia idade me', 'data nascimento', 'unstack')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por sexo')

plota_fig(sinasc,'PESO',['DTNASC','SEXO'],'mean','mdia peso beb', 'data nascimento', 'unstack')
plt.savefig('./output/figs/'+max_data+'/mdia peso beb por sexo')

plota_fig(sinasc,'APGAR1','GESTACAO','mean','apgard mdio', 'gestao', 'sort')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por data')

sinasc = pd.read_csv('./input/SINASC_RO_2019_JUN.csv')

max_data = sinasc.DTNASC.max()[:7]
print(max_data)

os.makedirs('./output/figs/'+max_data)

plota_fig(sinasc,'IDADEMAE','DTNASC','mean','mdia idade me por data', 'data nascimento')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por data')

plota_fig(sinasc,'PESO','ESCMAE','median','peso mediano beb', 'escolaridade me', 'sort')
plt.savefig('./output/figs/'+max_data+'/peso mediano por escolaridade me')

plota_fig(sinasc,'IDADEMAE',['DTNASC','SEXO'],'mean','mdia idade me', 'data nascimento', 'unstack')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por sexo')

plota_fig(sinasc,'PESO',['DTNASC','SEXO'],'mean','mdia peso beb', 'data nascimento', 'unstack')
plt.savefig('./output/figs/'+max_data+'/mdia peso beb por sexo')

plota_fig(sinasc,'APGAR1','GESTACAO','mean','apgard mdio', 'gestao', 'sort')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por data')

sinasc = pd.read_csv('./input/SINASC_RO_2019_DEZ.csv')

max_data = sinasc.DTNASC.max()[:7]
print(max_data)

os.makedirs('./output/figs/'+max_data)

plota_fig(sinasc,'IDADEMAE','DTNASC','mean','mdia idade me por data', 'data nascimento')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por data')

plota_fig(sinasc,'PESO','ESCMAE','median','peso mediano beb', 'escolaridade me', 'sort')
plt.savefig('./output/figs/'+max_data+'/peso mediano por escolaridade me')

plota_fig(sinasc,'IDADEMAE',['DTNASC','SEXO'],'mean','mdia idade me', 'data nascimento', 'unstack')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por sexo')

plota_fig(sinasc,'PESO',['DTNASC','SEXO'],'mean','mdia peso beb', 'data nascimento', 'unstack')
plt.savefig('./output/figs/'+max_data+'/mdia peso beb por sexo')

plota_fig(sinasc,'APGAR1','GESTACAO','mean','apgard mdio', 'gestao', 'sort')
plt.savefig('./output/figs/'+max_data+'/mdia idade me por data')
```
