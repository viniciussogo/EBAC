Para isto, conduziu um estudo com alguns clientes, comprovando suas rendas através de olerites e outros documentos, e pretende construir um modelo preditivo para esta renda com base em algumas variáveis que já possui em seu banco de dados.

1) Avalie a estrutura de correlação das variáveis quantitativas através de um gráfico de "matriz de dispersão" e através da avaliação gráfica da matriz de correlações. Veja se você identifica algum padrão interessante ou que te faça sentido.

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
dados = pd.read_csv('previsao_de_renda.csv')
dados.head()
````

````python
variaveis_quantitativas = dados[['idade', 'tempo_emprego', 'qt_pessoas_residencia', 'renda']]
````

````python
sns.pairplot(variaveis_quantitativas, diag_kind="kde", plot_kws={'alpha': 0.6})
plt.suptitle("Matriz de Dispersão das Variáveis Quantitativas", y=1.02)
plt.show()
````

````python
# Calcular a matriz de correlação
correlacoes = variaveis_quantitativas.corr()

# Criar um mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(correlacoes, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Matriz de Correlação das Variáveis Quantitativas")
plt.show()
````

4) Calcule o logaritmo da variável renda e repita a sequência de análise

````python
# Adicionar uma nova coluna com o logaritmo da renda
dados['log_renda'] = np.log(dados['renda'])

# Selecionar as variáveis quantitativas para incluir 'log_renda'
variaveis_quantitativas_log = dados[['idade', 'tempo_emprego', 'qt_pessoas_residencia', 'log_renda']]
````

````python
sns.pairplot(variaveis_quantitativas_log, diag_kind="kde", plot_kws={'alpha': 0.6})
plt.suptitle("Matriz de Dispersão com Log(Renda)", y=1.02)
plt.show()
````

````python
# Calcular a matriz de correlação
correlacoes_log = variaveis_quantitativas_log.corr()

# Criar um mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(correlacoes_log, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Matriz de Correlação com Log(Renda)")
plt.show()
````
