````python
#Importar os pacotes
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
dados = pd.read_csv('previsao_de_renda.csv')
dados.head()
````

````python
dados_qualitativos = dados[['posse_de_veiculo', 'posse_de_imovel', 'renda']]
dados_qualitativos.head()
````

````python
ax = sns.pointplot(x='posse_de_veiculo', y='renda', data=dados_qualitativos, dodge=True, ci=95)
````

````python
ax_1 = sns.pointplot(x='posse_de_imovel', y='renda', data=dados_qualitativos, dodge=True, ci=95)
````

````python
ax_2 = sns.pointplot(x='posse_de_veiculo', y='renda', hue='posse_de_imovel', data=dados_qualitativos, dodge=True, ci=95)
````
