# Cálculo - Tarefa 02

1- Calcule de forma numérica a integral da função $x^3$ avaliada entre os pontos $1$ e $2$. Identifique o número de retângulos que você usou.  
2- Calcule essa área de forma analítica.

```python
import numpy as np
import matplotlib.pyplot as plt

# Função x³
f = lambda x: x**3

def calcula_area(a, b, func, frame_num):
    num_bins = frame_num + 1  # Quantidade de retângulos
    delta = (b - a) / num_bins  # Largura dos retângulos
    x_cols = np.linspace(a, b - delta, num_bins) + delta / 2  # Pontos centrais
    y_cols = func(x_cols)  # Altura dos retângulos
    return sum(delta * y_cols)

def animate(n, a=1, b=2, func=f, verdadeiro_valor=3.75):
    plt.cla()  # Limpa o gráfico anterior

    ax = plt.axes(xlim=(a, b), ylim=(0, b**3 + 1))  # Ajusta eixos

    # Plota a função
    x = np.linspace(a, b, 100)
    y = func(x)
    ax.plot(x, y, '-')

    # Definição dos retângulos
    num_bins = n + 1
    delta = (b - a) / num_bins
    x_cols = np.linspace(a, b - delta, num_bins) + delta / 2
    y_cols = func(x_cols)

    # Plota os retângulos
    bar = ax.bar(x_cols, y_cols, alpha=0.5, width=delta, edgecolor='blue')

    # Soma de Riemann (área aproximada)
    soma_riemann = calcula_area(a, b, func, n)

    # Erro
    erro = verdadeiro_valor - soma_riemann
    ax.set_title(f'Soma de Riemann = {soma_riemann:.4f} | Erro = {erro:.4f} | Retângulos: {num_bins}')

    return bar

animate(3)
plt.show()
```
